```python
import time
import sys
import os 

# -Infinity implementation
minf = -sys.maxsize-1
inf = sys.maxsize+1
     
#RECURSIVE_MATRIX_CHAIN
def RECURSIVEMATRIXCHAIN(p, i, j):  
    if i == j: 
        return 0
  
    mij = sys.maxsize+1 
  
    for k in range(i, j):   
        q = (RECURSIVEMATRIXCHAIN(p, i, k) + RECURSIVEMATRIXCHAIN(p, k + 1, j) + p[i-1] * p[k] * p[j])  
        if q < mij: 
            mij = q      
    return mij 

#MEMOIZED_MATRIX_CHAIN
def MEMOIZEDMATRIXCHAIN(p):
    n = len(p) -1
    m = [[-1 for x in range(n+1)] for x in range(n+1)]
    s = [[-1 for x in range(n+1)] for x in range(n+1)]
    for i in range (1, n+1):
        for j in range (i , n+1):
            m[i][j] = inf
    LOOKUPCHAIN(m,p,1,n,s)
    return m, s

def LOOKUPCHAIN(m,p,i,j,s):
    if m[i][j] < inf:
        return m[i][j]
    if i == j:
        q=0
    else:
        q = inf
        for k in range (i, j):
            val = LOOKUPCHAIN(m,p,i,k,s)+ LOOKUPCHAIN(m,p,k+1,j,s) + p[i-1] * p[k] * p[j]
            if val < q:
                q = val
                s[i][j] = k
    m[i][j]=q
    return m[i][j]  

#MATRIX_CHAIN_ORDER
def MATRIXCHAINORDER(p):
    n = len(p)
    #length = len(p) 
    m = [[-1 for x in range(n)] for x in range(n)]
    s = [[-1 for x in range(n)] for x in range(n)]
    for i in range(1,n):
        m[i][i]=0
    for l in range(2, n):
        for i in range(1, n - l + 1):
            j = i + l - 1
            q = inf
            for k in range(i, j):
                val = m[i][k] + m[k + 1][j] + p[i - 1]*p[k]*p[j]
                if q>val:
                    q=val
                    s[i][j] = k
            m[i][j] = q
    return m, s

def PRINTPARENTHESIS(s, i, j):
    if i == j:
        print('A{}'.format(i), end='')
        return 0 
    k = s[i][j]
 
    print('(', end='')
    PRINTPARENTHESIS(s, i, k)
    PRINTPARENTHESIS(s, k + 1, j)
    print(')', end='')

# Input from user
n = int(input('Enter number of args: '))
lenval = n
p = []
for i in range (1, n+1):
    pelement = input('Enter the value of p{} in '.format(i-1))
    p.append(int(pelement))

#Start time
beginTime = time.time()
#print output 
a = RECURSIVEMATRIXCHAIN(p, 1, n-1)
endTime = time.time()
print("Recursive Matrix Chain: ")
print (a)
# Print time taken in microseconds
print(str(1000000*(endTime - beginTime))) 

#Start time
beginTime = time.time()
#print output 
m,s = MEMOIZEDMATRIXCHAIN(p)
endTime = time.time()
print("Memoized Matrix Chain: ")
print (m[1][n-1])
PRINTPARENTHESIS(s, 1, n-1)
# Print time taken in microseconds
tym = str(1000000*(endTime - beginTime))
print (' \n', end = '')
print(tym)

#Start time
beginTime = time.time()
#print output 
m,s = MATRIXCHAINORDER(p)
endTime = time.time()
print("Bottom-up Matrix Chain: ")
print (m[1][n-1])
PRINTPARENTHESIS(s, 1, n-1)
# Print time taken in microseconds
tym = str(1000000*(endTime - beginTime))
print ('\n', end = '')
print(tym)

#Sample inputs
#5, 2, 3, 10, 5, 4, 20 
#6, 30, 35, 15, 5, 10, 20, 25 
#3, 6, 8, 11, 14, 19, 22, 26, 27, 31, 36, 39, 42, 46
#10,10,10,15,17,19,25,25,27,28,35,37,37,39,40,50,53,58,56,64
#2,5,23,21,4,5,6,9,33,11,4,30,4,22,6,9,44,5,3,22,5,7
```
